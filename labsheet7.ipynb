```python
#1
f = open("file1.txt", "w")
f.write("Hello, this is a sample text.")
f.close()

#2
f = open("file1.txt", "r")
print(f.read())
f.close()

#3
f = open("file1.txt", "r")
for line in f:
    print(line.strip())
f.close()

#4
f = open("file1.txt", "r")
words = f.read().split()
print("Number of words:", len(words))
f.close()

#5
f = open("file1.txt", "r")
lines = f.readlines()
chars = sum(len(line) for line in lines)
print("Lines:", len(lines))
print("Characters:", chars)
f.close()

#6
f = open("file1.txt", "r")
lines = f.readlines()
for i in range(1, len(lines), 2):
    print(lines[i].strip())
f.close()

#7
f = open("file2.txt", "w")
n = int(input("Enter number of lines: "))
for i in range(n):
    line = input("Enter line: ")
    f.write(line + "\n")
f.close()

#8
f = open("file1.txt", "a")
f.write("\nThis is newly appended content.")
f.close()

#9
f1 = open("file1.txt", "r")
f2 = open("file_copy.txt", "w")
f2.write(f1.read())
f1.close()
f2.close()

#10
f = open("file1.txt", "r")
data = f.read().upper()
f.close()
print(data)

#11
f = open("file1.txt", "r")
word = input("Enter word to search: ")
found = False
for line in f:
    if word in line:
        found = True
        break
print("Word found" if found else "Word not found")
f.close()

#12
f = open("file1.txt", "r")
word = input("Enter word to count: ")
count = f.read().split().count(word)
print("Occurrences:", count)
f.close()

#13
f = open("numbers.txt", "r")
nums = [int(x) for x in f.read().split()]
print("Sum:", sum(nums))
print("Average:", sum(nums) / len(nums))
f.close()

#14
f = open("file1.txt", "r")
lines = f.readlines()
longest = max(lines, key=len)
print("Longest line:", longest.strip())
f.close()

#15
f = open("file1.txt", "r")
text = f.read().replace("oldword", "newword")
f.close()
f = open("file1.txt", "w")
f.write(text)
f.close()

#16
f = open("file1.txt", "r")
lines = f.readlines()
print("First line:", lines[0].strip())
print("Last line:", lines[-1].strip())
f.close()

#17
f = open("file1.txt", "r")
lines = f.readlines()
f.close()
f = open("file1.txt", "w")
for line in lines:
    if line.strip():
        f.write(line)
f.close()

#18
f1 = open("file1.txt", "r")
words = set(f1.read().split())
f1.close()
f2 = open("unique.txt", "w")
for w in words:
    f2.write(w + "\n")
f2.close()

#19
f1 = open("file1.txt", "r")
words = sorted(f1.read().split())
f1.close()
f2 = open("sorted_words.txt", "w")
f2.write(" ".join(words))
f2.close()

#20
f1 = open("fileA.txt", "r")
f2 = open("fileB.txt", "r")
f3 = open("merged.txt", "w")
f3.write(f1.read() + "\n" + f2.read())
f1.close()
f2.close()
f3.close()

#21
import csv
with open("students.csv", "w", newline='') as f:
    w = csv.writer(f)
    w.writerow(["Name", "Marks"])
    w.writerow(["Amit", 85])
    w.writerow(["Riya", 90])

#22
import csv
with open("students.csv", "r") as f:
    r = csv.reader(f)
    for row in r:
        print(row)

#23
import csv
with open("students.csv", "r") as f:
    r = csv.reader(f)
    next(r)
    marks = [int(row[1]) for row in r]
print("Average:", sum(marks)/len(marks))

#24
import csv
with open("students.csv", "r") as f:
    r = list(csv.reader(f))
    print("Rows:", len(r))
    print("Columns:", len(r[0]))

#25
import csv
with open("students.csv", "a", newline='') as f:
    w = csv.writer(f)
    w.writerow(["Karan", 88])

#26
import csv
with open("students.csv", "r") as f1, open("students_copy.csv", "w", newline='') as f2:
    for row in f1:
        f2.write(row)

#27
import csv
with open("students.csv", "r") as f:
    r = csv.DictReader(f)
    for row in r:
        print(row["Name"], row["Marks"])

#28
import csv
with open("students.csv", "r") as f:
    r = csv.DictReader(f)
    max_marks = 0
    top_student = ""
    for row in r:
        if int(row["Marks"]) > max_marks:
            max_marks = int(row["Marks"])
            top_student = row["Name"]
print("Topper:", top_student, "with", max_marks)

#29
import csv
name = input("Enter student name: ")
with open("students.csv", "r") as f:
    r = csv.DictReader(f)
    for row in r:
        if row["Name"] == name:
            print(row)
            break

#30
import csv
files = ["file1.csv", "file2.csv", "file3.csv"]
with open("combined.csv", "w", newline='') as out:
    writer = csv.writer(out)
    for file in files:
        with open(file, "r") as f:
            for row in csv.reader(f):
                writer.writerow(row)
```
