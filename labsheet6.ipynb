```python
#1
s = {1, 2, 3, 4}
print(s)

#2
s = {10, "Hello", 3.5, True}
print(s)

#3
s = {1, 2, 3}
s.add(4)
print(s)

#4
s = {1, 2}
s.update([3, 4, 5])
print(s)

#5
s = {1, 2, 3, 4}
s.remove(3)
print(s)

#6
s = {10, 20, 30}
s.discard(20)
print(s)

#7
s = {1, 2, 3}
s.clear()
print(s)

#8
s = {5, 10, 15, 20}
print(10 in s)

#9
s = {1, 2, 3, 4, 5}
print(len(s))

#10
s = {10, 20, 30}
for i in s:
    print(i)

#11
s1 = {1, 2, 3}
s2 = {3, 4, 5}
print(s1 | s2)

#12
s1 = {1, 2, 3}
s2 = {2, 3, 4}
print(s1 & s2)

#13
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5}
print(s1 - s2)

#14
s1 = {1, 2, 3}
s2 = {3, 4, 5}
print(s1 ^ s2)

#15
s1 = {1, 2}
s2 = {1, 2, 3, 4}
print(s1.issubset(s2))

#16
s1 = {1, 2, 3, 4}
s2 = {2, 3}
print(s1.issuperset(s2))

#17
s1 = {1, 2, 3}
s2 = {4, 5, 6}
print(s1.isdisjoint(s2))

#18
s1 = {1, 2, 3, 4, 5}
s2 = {2, 3}
s1.difference_update(s2)
print(s1)

#19
from itertools import chain, combinations
s = {1, 2, 3}
power_set = list(chain.from_iterable(combinations(s, r) for r in range(len(s)+1)))
print(power_set)

#20
A = {1, 2}
B = {3, 4}
cartesian = {(a, b) for a in A for b in B}
print(cartesian)

#21
lst = [1, 2, 2, 3, 4, 4, 5]
distinct = set(lst)
print(len(distinct))

#22
lst = [1, 2, 2, 3, 4, 4, 5]
duplicates = {x for x in lst if lst.count(x) > 1}
print(duplicates)

#23
lst = [1, 2, 2, 3, 4, 4, 5]
unique = list(set(lst))
print(unique)

#24
s = set("python")
print(s)

#25
s1 = set("apple")
s2 = set("grape")
print(s1 & s2)

#26
sentence = "this is a tree"
words = set(sentence.split())
print(words)

#27
str1 = "listen"
str2 = "silent"
print(set(str1) == set(str2))

#28
fs = frozenset([1, 2, 3, 4])
print(fs)


#29
s = {5, 10, 2, 8, 3}
print(max(s))
print(min(s))

#30
A = {1, 2, 3}
B = {3, 4, 5}
print("Union:", A | B)
print("Intersection:", A & B)
print("Difference:", A - B)
print("Symmetric Difference:", A ^ B)
```
